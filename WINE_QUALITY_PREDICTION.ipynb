```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
```

DATA ANALYSIS AND VISUALIZATION :  

```python
# statistical measures of DF
df.describe()
```

```python
# number of values for each quality
sns.catplot(x='quality',data=df,kind='count',color='green')
```

```python
# volatile acidity vs quality
plot=plt.figure(figsize=(5,5))
sns.barplot(x='quality',y='volatile acidity',data=df)
```

CORRELATION BETWEEN DF COLUMNS:

1. POSITIVE CORRELATION - IF BOTH VALUES INCREASES OR DECREASES SIMULTANEOUSLY

2. NEGATIVE CORRELATION - IF ONE VALUE INCREASES OTHER DECREASES AND VICE VERSA

```python
correlation = df.corr()
```

```python
# constructing a heatmap to understand the correlation between columns
plt.figure(figsize=(10,10))
sns.heatmap(correlation,cbar=True,square=True,fmt='.1f',annot=True,annot_kws={'size':8},cmap='Blues')
```

HERE WE CAN ANALYSE WHICH COLUMNS ARE POSITIVELY OR NEGATIVELY CORRELATED WITH OUR TARGET/LABEL COLUMN (QUALITY)

```python
# DATA PREPROCESSING
# separating the data and label
X= df.drop('quality',axis=1)
print(X)
```

LABEL BINARIZATION :

```python
Y= df['quality'].apply(lambda y_value :1 if y_value>=7 else 0)
print(Y)
```

TRAIN TEST SPLITTING

```python
X_train, X_test, Y_train, Y_test = train_test_split(X,Y,test_size=0.2,random_state=2)
```

```python
print(X.shape,X_train.shape,X_test.shape)
```

MODEL TRAINING :

RANDOM FOREST CLASSIFIER -- ENSEMBLE MODEL OF DECISION TREES

```python
model = RandomForestClassifier()
```

```python
# training the model
model.fit(X_train,Y_train)
```

MODEL EVALUATION:

```python
# accuracy score on test data
X_test_prediction= model.predict(X_test)
test_data_accuracy = accuracy_score(X_test_prediction,Y_test)
print(" accuracy on testing data :", test_data_accuracy)
```

HERE WE ARE GETTING ACCURACY SCORE OF 92 PERCENT WHICH MEANS OUR MODEL IS PREDICTING VERY GOOD.

GENERALLY ACCURACY SCORE > 75 PERCENT IS CONSIDERED GOOD FOR PREDICTION.

BUILDING A PREDICTIVE SYSTEM:

```python
input_data=(7.4,0.66,0.0,1.8,0.075,13.0,40.0,0.9978,3.51,0.56,9.4) # it should predict 0

# changing the input data to a numpy array
input_data_as_numpy_array = np.asarray(input_data)

# reshape the data as we are predicting the label for only one instance
input_data_reshaped=input_data_as_numpy_array.reshape(1,-1)

prediction = model.predict(input_data_reshaped)
print(prediction)

if prediction[0]==1 :
  print(" Good quality wine")
else :
  print("Bad quality wine")
```

OUR PREDICTIVE SYSTEM IS PREDICTING CORRECTLY, HENCE OUR MODEL IS WORKING FINE.
